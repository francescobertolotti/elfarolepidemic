```python
import matplotlib.pyplot as plt
class learning():
    def __init__(self):

        self.q_table = np.zeros(2)
        self.update_rate = 0.01
        self.explor_rate = 0.1
        self.epochs = 10000

        self.q0_ts = []
        self.q1_ts = []
        self.decisions = []


    def get_reward(self, my_action):
        if my_action == 0:
            if np.random.random() < 0.5:
                return 0.75
            else:
                return 0.25
        else:
            return 0.5
        
    def plot_decisions(self):
        plt.plot(self.decisions)
        plt.title('decisions')
        plt.xlabel('epoch')
        plt.ylabel('decision')
        plt.show()

    def plot_results(self):
        plt.plot(self.q0_ts)
        plt.plot(self.q1_ts)
        plt.title('q_table')
        plt.xlabel('epoch')
        plt.ylabel('q')
        plt.legend(['q0', 'q1'])
        plt.show()



    def learn(self):
        for e in range(self.epochs):
                
            if np.random.random() < self.explor_rate:
                my_action = np.random.randint(2)
            else:
                if len(self.decisions) == 0:
                    my_action = np.random.randint(2)
                else:
                    my_action = np.argmax(self.q_table)
            
            self.decisions.append(my_action)
            reward = self.get_reward(my_action)
            self.q_table[my_action] = self.q_table[my_action] * (1 - self.update_rate) + self.update_rate * reward

            self.q0_ts.append(self.q_table[0])
            self.q1_ts.append(self.q_table[1])


        print(self.q_table)


    

mod = learning()
mod.learn()
mod.plot_results()
mod.plot_decisions()
```
